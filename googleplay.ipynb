```python
feature_columns = []
for col in ['Rating', 'Rating Count', 'Installs','Maximum Installs','Editor Choice']:
    if col in df_cleaned2.columns:
        feature_columns.append(col)
```

```python
def clean_installs(x):
    """
    Removes commas and plus signs from a string and converts it to an integer.
    If x is not a string, it returns the value as-is.
    """
    if isinstance(x, str):
        # Remove commas and plus signs
        cleaned = x.replace(',', '').replace('+', '')
        try:
            return int(cleaned)
        except ValueError:
            # If conversion fails, return NaN
            return np.nan
    return x

# Apply cleaning to the 'Installs' column (or any similar column)
if 'Installs' in df_cleaned2.columns:
    df_cleaned2['Installs'] = df_cleaned2['Installs'].apply(clean_installs)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X_appname = tfidf_vectorizer.fit_transform(df_cleaned2['App Name'].fillna(''))
```

```python
if 'Category' in df_cleaned2.columns:
    le = LabelEncoder()
    df_cleaned2['Category_encoded'] = le.fit_transform(df_cleaned2['Category'])
    # Scale and convert to sparse matrix
    X_category = scaler.fit_transform(df_cleaned2[['Category_encoded']])
    X_category_sparse = csr_matrix(X_category)
    # Combine numeric, app name, and category features
    X_final = hstack([X_numeric_sparse, X_appname, X_category_sparse])
else:
    # Combine numeric features and app name features
    X_final = hstack([X_numeric_sparse, X_appname])

print("Final feature matrix shape:", X_final.shape)
```

```python
n_clusters = 2  # For example, suspicious vs. not suspicious
batch_size = 1000  # Adjust as needed

model = MiniBatchKMeans(n_clusters=n_clusters, batch_size=batch_size, random_state=42)
model.fit(X_final)

# Assign cluster labels
df_cleaned2['cluster'] = model.labels_
```

```python
silhouette_avg = silhouette_score(X_final, model.labels_)
print("Silhouette Score:", silhouette_avg)
```
